```stata
use data/hdefa_dist2012_2020, clear
```

```stata
qui {
drop if control<1
collapse (sum) efdeexc, by(year control)
reshape wide efdeexc, i(year) j(control)
label var efdeexc3 "for-profit    "
label var efdeexc2 "nonprofit (private)"
label var efdeexc1 "nonprofit (public)"  
}
```

```stata
tw (connect efdeexc3 efdeexc2 efdeexc1 year if year>2011 & year<2020, sort ///
legend(pos(6) size(small) order(1 4 2 3)) xlabel(2011(2)2020, ang(45) labsize(medium)) ///
ytitle("online enrollment" " " " ", size(medium)) ylabel(, format(%10.0fc) labsize(medium)) ///
xtitle("")  note("") scheme(plotplain))
```
