```stata
quietly {
use data/d_OPM_database_paperJN.dta, clear
set more off 
set line 255
}
```

```stata
estimates clear

qui	eststo m1VC_or_PE_dum: logit TP_recruiting VC_or_PE_dum if LMS!=1
	
qui	eststo m2VC_or_PE_dum: logit Rev_share_w_TP VC_or_PE_dum LMS
			
qui	eststo m4VC_or_PE_dum: logit TP_chgs_per_head VC_or_PE_dum LMS
	
qui	eststo m5VC_or_PE_dum: logit no_rs_ph VC_or_PE_dum LMS
	
qui	eststo m3VC_or_PE_dum: logit invisible_1 VC_or_PE_dum if LMS!=1
	
qui	eststo m6VC_or_PE_dum: logit Captivity_AR_INDEF VC_or_PE_dum 
	
qui	eststo m7VC_or_PE_dum: logit Expansion VC_or_PE_dum LMS
	
```

```stata
*Table 4*
*Relationship between Third-Party Financing Structure and Partnership Features *
esttab m1VC_or_PE_dum m2VC_or_PE_dum m4VC_or_PE_dum m5VC_or_PE_dum ///
	m3VC_or_PE_dum m7VC_or_PE_dum m6VC_or_PE_dum, ///
			se(%8.2f) b(%8.2f) star(+ 0.10 * 0.05 ** 0.01 *** 0.001) ///
			label title("Relationship between Third-Party Financing Structure and Partnership Features ") nolz

/*
Code below will export regression table in .rtf/Word file fomrmat.
Image of regression table used in paper can be viewed below.

esttab m1VC_or_PE_dum m2VC_or_PE_dum m4VC_or_PE_dum m5VC_or_PE_dum ///
	m3VC_or_PE_dum m7VC_or_PE_dum m6VC_or_PE_dum ///
		using OPM_final.rtf, ///
		se(%8.2f) b(%8.2f) star(+ 0.10 * 0.05 ** 0.01 *** 0.001) ///
		label title("Relationship between Third-Party Financing Structure and Partnership Features ") nolz replace	
*/
```
