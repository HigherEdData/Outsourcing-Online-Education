```stata
use data/d_OPM_database_paperJN.dta, clear
set more off 
set line 220
qui drop if LMS==1
```

```stata
*****Did not do robust std errors*
*Estimates presented in Sociology of Ed. with robust standard errors and linear probability modeling*
estimates clear
qui eststo m1VC_or_PE_dum: reg TP_recruiting VC_or_PE_dum if LMS!=1, vce(robust)
	
qui eststo m2VC_or_PE_dum: reg Rev_share_w_TP VC_or_PE_dum if LMS!=1, vce(robust)
			
qui eststo m4VC_or_PE_dum: reg TP_chgs_per_head VC_or_PE_dum if LMS!=1, vce(robust)
	
qui eststo m5VC_or_PE_dum: reg no_rs_ph VC_or_PE_dum if LMS!=1, vce(robust)
	
qui eststo m3VC_or_PE_dum: reg invisible_1 VC_or_PE_dum if LMS!=1, vce(robust)
	
qui eststo m6VC_or_PE_dum: reg Captivity_AR_INDEF VC_or_PE_dum if LMS!=1, vce(robust)
	
qui eststo m7VC_or_PE_dum: reg Expansion VC_or_PE_dum if LMS!=1, vce(robust)	


esttab m1VC_or_PE_dum m2VC_or_PE_dum m4VC_or_PE_dum m5VC_or_PE_dum ///
	m3VC_or_PE_dum m7VC_or_PE_dum m6VC_or_PE_dum, ///
			se(%8.2f) b(%8.2f) star(+ 0.10 * 0.05 ** 0.01 *** 0.001) ///
			label title("Relationship between Third-Party Financing Structure and Partnership Features ") nolz	
```
